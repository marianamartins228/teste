```python
#importação de bibliotecas
import random
import matplotlib.pyplot as plt

#gerar uma lista com 10 numeroos aleatórios entre 1 e 100
dados = [random.randint(1,100) for _ in range(10)]
#Exibir a lista gerada
print("Lista de dados:", dados)

#plotar os dados
plt.plot(dados, marker='o') # marcador em cada ponto para melhor visualização
plt.title("Gráfico de números aleatórios")
plt.xlabel("Índice")
plt.ylabel("Valor")
plt.grid(True)
plt.show()
```
